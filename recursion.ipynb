Lambda Functions:
A lambda function, also known as an anonymous function, is a small, unnamed function that can have any number of arguments but can only have one expression. Lambda functions are often used for short, simple operations where a full function definition would be overly verbose.
syntax
                lambda arguments: expression


```python
# calculate the number of percentage value
# Celsius to Fahrenheit:
# Fahrenheit (°F) = (Celsius (°C) × 9/5) + 32

# Fahrenheit to Celsius:
# Celsius (°C) = (Fahrenheit (°F) - 32) × 5/9

print("Choices for ")
print("1.Celsius to Fahrenheit: ")
print("2.Fahrenheit to Celsius")

try:
    ch=int(input('Your choices'))
    if ch==1:
        print("Enter the temperature in Celsius : ")
        Celsius=float(input("Celsius :"))
        con=lambda  Celsius:Celsius*9/5 + 32
        re=con(Celsius)
        print(re)

    else:
        print("Enter the temperature in Celsius : ")
        Fahrenheit=float(input("Fahrenheit :"))
        con=lambda  Fahrenheit:Fahrenheit-32*5/9
        re=con(Fahrenheit)
        print(re)
except ValueError:
    print("Invalide code")
```

```python
def Celsius(c):
    return lambda Celsius:Celsius*9/5 + 32
def Fahrenheit(f):
    return lambda  f:f-32*5/9
print("Choices for ")
print("1.Celsius to Fahrenheit: ")
print("2.Fahrenheit to Celsius")
ch=int(input("Choices please"))
if ch==1:
    print("\nEnter the temperature in celsius:")
    temp=float(input("Temperature please"))
    re=Celsius(temp) #here the temp is pass in fun 1 then we get Celsius ans in re again 
    final_lambda=re(temp)# here also then pass the temp again ans fun re
    print("\nTemperature :",temp , "Celsius (°C) => ",final_lambda ," Fahrenheit (°F) ")
elif ch ==2:
    print("\nEnter the temperature in Fahrenheit (°F) :")
    temp=float(input("Temperature please"))
    re=Fahrenheit(temp)
    final_lambda=re(temp)# here also then pass the temp again ans fun re
    print("\nTemperature :",temp , "Fahrenheit (°F) => " ,final_lambda, "Celsius (°C) ")
```
